```python
import requests
import pickle
import gzip
import urllib.request
import json
from bs4 import BeautifulSoup
```

```python
def breakIn(string):
    
    soup = BeautifulSoup(string, 'lxml')
    spans = soup.find_all('span')
    
    item_palavra = None
    infos = {}
    
    for span in spans:
        if span['class'][0] == "item-palavra":
            item_palavra = span.text
        else:
            infos[span['class'][0]] = span.text
    
    return item_palavra, infos
```

```python
def requestWord(old = None):
    
    retorno = {}
    
    alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q',
               'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    
    for letra in alfabeto:
        
        string = old if old != None else ""
        string += letra
        
        urlBase = "http://www.academia.org.br/ajax/abl/buscar-palavras?form=vocabulario&palavra=" + string + "%25"
        
        while True:
            try:
                req = urllib.request.Request(urlBase)
                opener = urllib.request.build_opener()
                f = opener.open(req)
                words_json = json.loads(f.read())
                if "rows" not in words_json:
                    raise ValueError('Request mal feito.')
                break
            except:
                print("Error: "+string)
                pass
            
        
        tamanho = len(words_json['rows'])
        
        for i in range(tamanho):
            item_palavra, infos = breakIn(words_json['rows'][i]['palavra'])
            retorno[item_palavra] = infos
        
        if tamanho >= 100:
            retorno = {**retorno, **requestWord(string)}
            
    print(string + " retornou " + str(len(retorno)))
    
    return retorno
    
```

```python
dicionario = requestWord()
print("Quantidade de Verbetes: " + str(len(dicionario)))

#store the object
f = gzip.open('dicionario.pklz','wb')
pickle.dump(dicionario, f)
f.close()
```
